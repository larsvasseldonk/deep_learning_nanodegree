# Hyperparameters

In this lesson, we will cover the following topics:
- Key Types of Hyperparameters
- Hyperparameters used in RNN

By the end of the lesson, you'll be able to:
- Define the key types of hyperparameters
- Create intuition about hyperparameter starting values

Hyperparamters can be divided into two broad categories:
- Training (also called optimizer) hyperparameters:
  - Learning rate
  - Minibatch size
  - Training epochs
- Model hyperparameters:
  - Number of layers
  - Model specific parameters
 
## Learning rate

Learning rate is the single most important hyperparameter. That is because the learning rate has the biggest impact on the gradient descent of our function.

**Intuition**: We want to see the error descend regularly. If the error does not consistently descend, we may have a problem with our learning rate.



Adaptive learning rates change over time to faster find the global minima. 

A learning rate that is too large can settle at a local minimum. A learning rate that is too small can take too long to fit the function

## Minibatch Size

Minibatch size is an important hyperparameter that can impact the performance of your network and speed of training.

The most commonly used minibatch sizes are:

1
2
4
8
16
32
64
128
256

Intuition: Large minibatch sizes can increase the speed of training. However, they often get caught at local minima during the training process.

Too low could be too slow. Too large could be too computational heavy.

A minibatch allows you to parallelize the way the data flows through the model.

## Number of Training Iterations / Epochs
The number of training iterations is a hyperparameter we can optimize automatically using a technique called early stopping (also "early termination").

We can stop training if the validation hasn't improve in the past 10 iterations for example.

## Number of Hidden Units / Layers
"In practice, it is often the case that 3-layer neural networks will outperform 2-layer nets, but going even deeper (4,5,6-layer) rarely helps much more. This is in stark contrast to Convolutional Networks, where depth has been found to be an extremely important component for a good recognition system (e.g. on order of 10 learnable layers)." ~ Andrej Karpathy in https://cs231n.github.io/neural-networks-1/

Keep adding hidden units untill the validation loss stops decreasing.

## RNN Hyperparamters
Two main choices that we have to make for RNNs:
- choosing a **cell type**: so either a LSTM cell, GRU cell or a vanilla RNN cell
- how deep the model is: how many layers will be stacked
